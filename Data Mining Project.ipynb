# Splitting Dataset into Training and Testing Data

```python
X = df.iloc[:,1:31].values
Y = df.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

def build_models(X_train, y_train):
    if not isinstance(X_train, pd.DataFrame):
        X_train = pd.DataFrame(X_train)
    num_cols = X_train.columns 

    preprocess_lr = ColumnTransformer(
        [("num", Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler())
        ]), num_cols)],
        remainder="drop"
    )
    preprocess_tree = ColumnTransformer(
        [("num", SimpleImputer(strategy="median"), num_cols)],
        remainder="drop"
    )

    # Models
    log = Pipeline([
        ("prep", preprocess_lr),
        ("clf", LogisticRegression(max_iter=1000, class_weight="balanced"))
    ])
    tree = Pipeline([
        ("prep", preprocess_tree),
        ("clf", DecisionTreeClassifier(max_depth=6, random_state=42, class_weight="balanced"))
    ])
    forest = Pipeline([
        ("prep", preprocess_tree),
        ("clf", RandomForestClassifier(n_estimators=400, random_state=42, class_weight="balanced"))
    ])

    # Fit (safe even if X_train has NaNs)
    log.fit(X_train, y_train)
    tree.fit(X_train, y_train)
    forest.fit(X_train, y_train)

    print("[] Logistic Regression, Decision Tree, and Random Forest fitted with imputation.")
    return log, tree, forest
```

```python
log, tree, forest = build_models(X_train, y_train)

print("Train accuracy (LogReg): ", log.score(X_train, y_train))
print("Train accuracy (Tree):   ", tree.score(X_train, y_train))
print("Train accuracy (Forest): ", forest.score(X_train, y_train))

# Evaluate on your held-out test set:
from sklearn.metrics import roc_auc_score, classification_report
for name, model in [("LogReg", log), ("Tree", tree), ("Forest", forest)]:
    proba = model.predict_proba(X_test)[:, 1]
    pred  = (proba >= 0.5).astype(int)
    print(f"\n{name}  Test ROC-AUC: {roc_auc_score(y_test, proba):.4f}")
    print(classification_report(y_test, pred, target_names=["Benign","Malignant"]))
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

num_preprocess = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", num_preprocess, num_cols)
    ],
    remainder="drop" 
)
```

```python
if {"perimeter_mean", "radius_mean"}.issubset(df.columns):
    df["shape_irregularity_index"] = df["perimeter_mean"] / df["radius_mean"]

if {"texture_worst", "texture_mean"}.issubset(df.columns):
    df["texture_roughness_score"] = df["texture_worst"] / df["texture_mean"]

if {"radius_worst", "radius_mean"}.issubset(df.columns):
    df["radius_severity_ratio"] = df["radius_worst"] / df["radius_mean"]

print("\nNew engineered features added:")
print([c for c in df.columns if c.endswith("_index") or "score" in c or "severity" in c])
```

### Data Reduction & Feature Selection

```python
from sklearn.feature_selection import mutual_info_classif

corr = df[num_cols].corr()

corr_abs = corr.abs()
upper = corr_abs.where(np.triu(np.ones(corr_abs.shape), k=1).astype(bool))
high_corr_pairs = upper.stack().sort_values(ascending=False)
print("\nTop highly correlated feature pairs:")
print(high_corr_pairs.head(10))

X_fs = df[num_cols]
y_fs = df["diagnosis"]

mi_scores = mutual_info_classif(X_fs, y_fs, random_state=42)
mi_series = pd.Series(mi_scores, index=num_cols).sort_values(ascending=False)
print("\nTop features by Mutual Information:")
print(mi_series.head(15))

selected_features = mi_series.head(15).index.tolist()
print("\nSelected top 15 features for modeling:")
print(selected_features)
```

### Final Preprocessed Dataset

```python
engineered_cols = [c for c in df.columns if c in [
    "shape_irregularity_index",
    "texture_roughness_score",
    "radius_severity_ratio"
]]
final_features = selected_features + engineered_cols

final_features = list(dict.fromkeys(final_features))

print("\nFinal feature set used for modeling:")
print(final_features)

X = df[final_features]
y = df["diagnosis"]

print("\nX shape:", X.shape)
print("y shape:", y.shape)
```

# Coding Part - Assignment 5

### Train-Test Split 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    stratify=y,
    random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

### Preprocessing Pipeline - Imputation + Scaling

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

numeric_features = X_train.columns  # all are numeric in WDBC

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features)
    ],
    remainder="drop"
)
```

### Logistic Regresssion - Training & Evaluate

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix

log_reg_clf = Pipeline(steps=[
    ("prep", preprocess),   # your imputer + scaler
    ("model", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        random_state=42
    ))
])

log_reg_clf.fit(X_train, y_train)
print("[] Trained Logistic Regression")

y_pred_log = log_reg_clf.predict(X_test)
y_proba_log = log_reg_clf.predict_proba(X_test)[:, 1]

acc_log = accuracy_score(y_test, y_pred_log)
auc_log = roc_auc_score(y_test, y_proba_log)
cm_log  = confusion_matrix(y_test, y_pred_log)

print("\n=== Logistic Regression ===")
print(f"Accuracy : {acc_log:.4f}")
print(f"ROC-AUC  : {auc_log:.4f}")
print("Confusion Matrix:\n", cm_log)
print("\nClassification Report:")
print(classification_report(y_test, y_pred_log, target_names=["Benign","Malignant"]))
```

### Decision Tree - Training & Evalute

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = Pipeline(steps=[
    ("prep", preprocess),
    ("model", DecisionTreeClassifier(
        max_depth=6,
        class_weight="balanced",
        random_state=42
    ))
])

tree_clf.fit(X_train, y_train)
print("[] Trained Decision Tree")

y_pred_tree = tree_clf.predict(X_test)
y_proba_tree = tree_clf.predict_proba(X_test)[:, 1]

acc_tree = accuracy_score(y_test, y_pred_tree)
auc_tree = roc_auc_score(y_test, y_proba_tree)
cm_tree  = confusion_matrix(y_test, y_pred_tree)

print("\n=== Decision Tree ===")
print(f"Accuracy : {acc_tree:.4f}")
print(f"ROC-AUC  : {auc_tree:.4f}")
print("Confusion Matrix:\n", cm_tree)
print("\nClassification Report:")
print(classification_report(y_test, y_pred_tree, target_names=["Benign","Malignant"]))
```

### Random Forest - Training & Evalute

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = Pipeline(steps=[
    ("prep", preprocess),
    ("model", RandomForestClassifier(
        n_estimators=400,
        class_weight="balanced",
        random_state=42
    ))
])

rf_clf.fit(X_train, y_train)
print("[] Trained Random Forest")

y_pred_rf = rf_clf.predict(X_test)
y_proba_rf = rf_clf.predict_proba(X_test)[:, 1]

acc_rf = accuracy_score(y_test, y_pred_rf)
auc_rf = roc_auc_score(y_test, y_proba_rf)
cm_rf  = confusion_matrix(y_test, y_pred_rf)

print("\n=== Random Forest ===")
print(f"Accuracy : {acc_rf:.4f}")
print(f"ROC-AUC  : {auc_rf:.4f}")
print("Confusion Matrix:\n", cm_rf)
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, target_names=["Benign","Malignant"]))
```
